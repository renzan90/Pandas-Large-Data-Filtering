```python
''' Here, we import the 'Mammal species.csv' file and print it out to check out its contents '''

the_mammal_file = pd.read_csv('Mammal species.csv')
print(the_mammal_file)
```

```python
''' We create a dataframe here from the 'Species' column in import csv file '''

all_species = pd.DataFrame(the_mammal_file['Species'])
print(all_species)
```

```python
'''

Here, we read the contents of the file 'Admittals-June-2018.csv', store it in 'all_data' and then print it

'''

all_data = pd.read_csv('Admittals-June-2018.csv', engine='python')
print(all_data)
```

```python
'''Here, we create a data frame in 'random_species' out of 'all_data', indexing it with the 'species' column and 
then printing it to check if the data has been successfully stored '''

random_species = pd.DataFrame(all_data['species'])
print(random_species)
```

```python
'''
Here, we concatenate the columns 'all_species' and 'random_species' to juxtapose them and then store them in
'combine1'. Then we save this to a new file titled 'first_combine.csv' 

'''

combine1 = pd.concat([all_species,random_species], axis=1, ignore_index=True)

print(combine1)

combine1.to_csv('first_combine.csv')
```

```python
''' 
First we initialize the empty list 'animal_list to store the names of all species in UK .Then we run a for loop
on the combined dataframe, within which we run another for loop for the list str_animals. The nested loops helps us
count the number of individuals in a given species. Also, the calling of the 'append' function for the list 'animal_list'
appending the species and the count helps us get a better view of species demographics in the U.K

'''

animal_list=[]
for animals in combine1[0]:
    count = 0
    print(animals)
    #animal_list.append([animals])
    for all in combine1[1]:
        if animals == all:
            count = count+1
    print(count)
    animal_list.append([animals, count])
print(animal_list)
```

```python
'''
We create a dataframe d3, where we add the list 'animal_list' with two elements separated as independent columns, 
namely, 'Species_Name' and 'Number Of Cases respectively'. 

However, here we start getting 'NA' where there are no values in the former column. We eliminate this error by 
using the dataframe function 'dropna()'. When we get the result, we display it.

'''

d3 = pd.DataFrame(animal_list, columns = [animals, count]) 
d3 = d3.dropna()
d3.columns = ["Species Name", "Number Of Cases"]
#d3.to_csv('result1.csv') 
print(d3)
```

```python
'''
Here, we find out the total number of cases by applying the 'sum' function on the d3 dataframe column 'Number Of Cases'

'''

Total = d3['Number Of Cases'].sum()
print(Total)
```

```python
'''
In this step, we accomplish three operations, namely, find the percentage of individual cases, rounding them up and
then converting them to integers. We again print the dataframe d3.
'''

d3['Percentage Of Total Cases'] = [x for x in round((d3['Number Of Cases']/Total)*100)]

d3['Percentage Of Total Cases'] = [int(to_integer) for to_integer in d3['Percentage Of Total Cases']]

print(d3)
```

```python
'''
Given that we want the elements in the final column of the d3 dataframe, we convert the values to string and wrap
them with single quotes and a percentage symbol to its right, by running a for loop through the last column.

For the column 'Number Of Cases', we convert the items to string and appending them with commas. 

For the first column in the dataframe, we reconvert the items to string despite them being strings already. Then we
flank them with single quotes and commas as demanded by the question format
'''

d3['Percentage Of Total Cases'] = [ "'" + str(x) + '%' + "'" for x in d3['Percentage Of Total Cases']]

d3['Number Of Cases'] = [str(cases) + "," for cases in d3['Number Of Cases']]

d3['Species Name'] = ["'" + str(species_name) + "'," for species_name in d3['Species Name']]

print(d3)
```

```python
'''
We stylize the above given result by calling the inbuilt dataframe functions like 'style'(which further calls 'hide_index')
and 'set_properties' functions that takes the three column headings as arguments. Further, we also align the text to the left by placing
another parameter beginning with '**{'text-align': 'left'}'.

'''

d3[["Species Name", "Number Of Cases", "Percentage Of Total Cases"]].style.hide_index().set_properties(["Species Name", "Number Of Cases", "Percentage Of Total Cases"], **{'text-align': 'left'}).set_table_styles([ dict(selector='th', props=[('text-align', 'left')] ) ]) 
```

```python
'''
And finally, we read the dataframe d3 to a csv file 'species-June2018.csv'.

'''
d3.to_csv('species-June2018.csv')
```

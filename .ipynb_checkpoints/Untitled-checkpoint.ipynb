```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import csv
```

```python
import random
```

```python
'''
Here, we import the provided file 'Admittals-June-2018.csv' which contains the information, most crucial to solve
the question, into a variable conveniently named as 'admittals_2018'.

'''

admittals_2018 = pd.read_csv('Admittals-June-2018.csv', engine='python')
print(admittals_2018)
```

```python
'''
We create a dataframe where we store the 'district' column of the 'admittals_2018' variable and print it out to see
if the contents are proper.

'''

admittals_district_dataframe = pd.DataFrame(admittals_2018['district'])
print(admittals_district_dataframe)
```

```python
'''
We simply declare and append to the following list the values taken from the above dataframe

'''

repeated_district_list = []

for x in admittals_district_dataframe['district']:
    repeated_district_list.append(x)
print(repeated_district_list)
```

```python
'''
Like one of the above steps, here too we import a csv file into a variable.

'''

UK_district_dataframe = pd.read_csv('Districts of the UK.csv')
print(UK_district_dataframe)
```

```python
'''
And here, we save it to a dataframe named 'districts_dataframe'.

'''

districts_dataframe = pd.DataFrame(UK_district_dataframe)
print(districts_dataframe)
```

```python
'''
The districts_dataframe is converted to a list by using the 'values' property and its subsequent 'tolist' function,
and stored in a list.

'''

districts_list = (districts_dataframe['Districts'].values.tolist())
print(districts_list)
```

```python
'''
This is a very important step. Here, we get the number of cases for each district lined up in a list in the same order
as that of the districts in the above given list.

'''

cases_per_district = []
for x in range(len(districts_list)):
    counter=0
    for y in range(len(repeated_district_list)):
        if districts_list[x] == repeated_district_list[y]:
            counter = counter + 1
    cases_per_district.append(counter)
print(cases_per_district)
```

```python
'''
We simply print the given regions of the U.K in the provided form, where they are repeated.

'''

regions = UK_district_dataframe['Regions'].values.tolist()
print(regions)
```

```python
'''
Here we create a nested list, from the lists 'cases_per_district', 'districts_list' and 'regions' by first zipping
them together with the 'zip' function into tuples and then using the 'map' function to map them into lists.

However, this reorders them as 'map objects', therefore, we again use the 'list' function to convert them into a list,

This list is very crucial for breaking down the third elements of each sublist, the 'region' sublist, thus not repeating
them more than once.

'''


cases_districts_and_regions = list(map(list, zip(cases_per_district, districts_list, regions)))
print(cases_districts_and_regions)
```

```python
'''
Here is another important portion of our code. This snippet breaks down the repeated list of 'regions' in the U.K
into one-time lists, without repeats.

After initializing the 'regions_one' list, we set a for loop, looping through the 'cases_districts_and_regions' list
and within that, the 'regions' list. 

The logic here is to compare the last element of the 'cases_district_and_regions' list to each element of the later.
If the condition gets satisfied, the region name gets appended to the 'regions_one' list.

'''


regions_one = []
for x in range(len(cases_districts_and_regions)):
    for y in range(len(regions)):
        if cases_districts_and_regions[x][2] == regions[y]:
            if regions[y] in regions_one:
                pass
            else:
                region = regions[y]
                regions_one.append(region)
print(regions_one)
```

```python
'''
In this another crucial step, we separate the number of cases for each region.

After creating the 'regional_cases' list, we run a for loop through the 'regions_one' list which will loop through
each region. A nested loop derived from the list 'cases_districts_and_regions' runs beneath. 

Here we compare whether each element in the first_list matches with the last element in the second list. If this
comparison satisfies the case, then we have a counter initialize right after the beginning of the first loop, which
we set to increment with the first value of the first element i.e. the number of cases. Then we append these values
to the regional_cases list.

'''


regional_cases = []
for y in range(len(regions_one)):
    counter = 0
    for x in range(len(cases_districts_and_regions)):
        if regions_one[y] == cases_districts_and_regions[x][2]:
            counter = counter + cases_districts_and_regions[x][0]
        else:
            pass
    regional_cases.append(counter)

print(regional_cases)            
```

```python
'''
In this step, we add all the elements in the 'regional_cases' list for carrying out further operations.

'''

total_cases = 0
for total in range(len(regional_cases)):
    total_cases = regional_cases[total] + total_cases
print(total_cases)
```

```python
'''
Here, we create a third and a also a crucial list called 'percentages', where we proportionalise the number of cases
per region into their respective percentages. As a result we get complex float values that we append to the 'percentages'
list.

'''


percentages = []
for z in range(len(regional_cases)):
    x = (regional_cases[z]/total_cases)*100
    percentages.append(x)
print(percentages)
```

```python
'''
Converting the three key lists, into their respective dataframes, namely, the 'regions_dataframe', the 'regional_cases
_dataframe', and the 'percentages_dataframe'.

This conversion will allow us to make changes to the data before finally writing it to the .csv file.

'''


regions_dataframe = pd.DataFrame(regions_one, columns = ['Region Name'])
regions_dataframe.dropna()
print(regions_dataframe)

regional_cases_dataframe = pd.DataFrame(regional_cases, columns = ["Number In Region"])
regional_cases_dataframe.dropna()
print(regional_cases_dataframe)

percentages_dataframe = pd.DataFrame(percentages, columns = ['Percentage Of Total Cases'])
percentages_dataframe.dropna()
print(percentages_dataframe)
```

```python
'''
Combine all three above-mentioned dataframes into a single 'main_dataframe' which will be later used for writing into
files.

'''

main_dataframe = pd.concat([regions_dataframe,regional_cases_dataframe, percentages_dataframe], axis=1, ignore_index=True)
main_dataframe.columns = ["Region Name", "Number In Region", "Percentage Of Total Cases"]
print(main_dataframe)
```

```python
Total = main_dataframe['Number In Region'].sum()
print(Total)
```

```python
'''
Here we round the values present in the 'Percentage Of Total Cases' column of the main_dataframe, which is a precur-
-sor to converting them to integer values.
'''

main_dataframe['Percentage Of Total Cases'] = [x for x in round((main_dataframe['Number In Region']/Total)*100)]

main_dataframe['Percentage Of Total Cases'] = [int(to_integer) for to_integer in main_dataframe['Percentage Of Total Cases']]

print(main_dataframe)
```

```python
'''
Since we are going to need the items in the third column as strings, first we convert them to strings than append
them with single quotes and percentage signs. 

We convert the second and first column elements too and then append them with commas and single quotes(and commas)
respectively

'''

main_dataframe['Percentage Of Total Cases'] = [ "'" + str(x) + '%' + "'" for x in main_dataframe['Percentage Of Total Cases']]

main_dataframe['Number In Region'] = [str(cases) + "," for cases in main_dataframe['Number In Region']]

main_dataframe['Region Name'] = ["'" + str(region_name) + "'," for region_name in main_dataframe['Region Name']]

print(main_dataframe)
```

```python
'''
We recreate the main_dataframe with converted values. However, in the current step, we also stylized the dataframe
by using the set_properties function.

'''

main_dataframe[["Region Name", "Number In Region", "Percentage Of Total Cases"]].style.hide_index().set_properties(["Region Name", "Number In Region", "Percentage Of Total Cases"], **{'text-align': 'left'}).set_table_styles([ dict(selector='th', props=[('text-align', 'left')] ) ]) 
```

```python
'''
Finally, we complete the task by writing the main_dataframe to a file titled 'region-June2018.csv'

'''

main_dataframe.to_csv('region-June2018.csv')
```

```python
'''
Since now we want to build a pie chart, we are going to need the column values in their original formats. In this 
step, we reconvert them to int formats for carrying out the further operation.

'''

Total = main_dataframe['Number In Region'].sum()
main_dataframe['Percentage Of Total Cases'] = [x for x in round((main_dataframe['Number In Region']/Total)*100)]

main_dataframe['Percentage Of Total Cases'] = [int(to_integer) for to_integer in main_dataframe['Percentage Of Total Cases']]

```

```python
'''
First, we create a list very obviously named as the 'initial_color_list' containing 17 colours. The 'final_colors'
list gets appended with colours from the previous list randomly in a range up to 102. For this, we call the random
module's 'choice' function.

We create 'x' and 'y' variables from the dataframe columns 'Region Name' and 'Percentage Of Total Cases' respectively.

Then we use the 'pie' function to plot the pie chart, where variable 'y' acts as criteria determinant and 'x' as label.

Finally, we save the image to a png file called 'piechart2.png'.

'''

initial_color_list = ['mediumturquoise', 'blueviolet', 'magenta', 'darksalmon', 'seagreen', 'crimson', 'darkblue', 'peru', 'linen', 'red', 'gold', 'azure', 'darkmagenta', 'salmon', 'darkorange', 'slategray', 'tomato', ]

final_colors = []
final_colors = [random.choice(initial_color_list) for i in range(0, 31)]

x = main_dataframe['Region Name']
y = main_dataframe['Percentage Of Total Cases']

plt.pie(y, labels = x, colors = final_colors, autopct = '%1.1f%%', shadow = True, startangle = 90, radius = 9)

plt.savefig('piechart2.png', bbox_inches='tight')
```
